```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.goodreads.com/quotes"
```

```python
data = requests.get(url)
```

```python
soup = BeautifulSoup(data.text)
```

```python
# Scraping the categories
options_lis = soup.find_all("li", attrs={"class": "greyText"})
option_text = []
for option_li in options_lis:
    option_info = option_li.text
    option = option_info.strip().split("\n")
    option_text.append(option[0])
```

```python
# asking for what category of quotes
print("Categories of quotes : ")
for i in range(len(option_text)):
    print(str(i+1) + " : " + option_text[i])

option = int(input("Enter the number : "))
```

```python
# getting the link of required quotes page
HREF = options_lis[option-1].find("a")['href']

link = url + HREF[7:]
data = requests.get(link)

soup = BeautifulSoup(data.text)

# and scraping the quotes and authors
quote_divs = soup.find_all("div", attrs={"class" : "quote"})
quote_text = []
quote_author = []

for quote in quote_divs:
    quote_div = quote
    quotetext = quote_div.find("div", attrs={"class" : "quoteText"})
    quote_info = quotetext.text
    quote_details = quote_info.strip().split("\n")
    text = quote_details[0]
    author = quote_details[3].strip()
    
    quote_text.append(text)
    quote_author.append(author)
```

```python
df = pd.DataFrame(list(zip(quote_text, quote_author)), 
               columns =['Quote', 'Author']) 
```
